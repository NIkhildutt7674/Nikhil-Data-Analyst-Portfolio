```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("sales_data_cleaned.csv")  # replace with your actual file path

# Quick sanity check
print(df.shape)
print(df[['Gender', 'Category', 'City', 'Total_Sales']].head())
print(df.isnull().sum())  # check for missing values in key columns
```

```python
print("\n[1] Independent T-Test: Male vs Female Spending")
male_sales = df[df['Gender'] == 'Male']['Total_Sales']
female_sales = df[df['Gender'] == 'Female']['Total_Sales']

t_stat, p_value = stats.ttest_ind(male_sales, female_sales, equal_var=False)

print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("Result: Reject H₀ → Significant difference in spending by gender.")
else:
    print("Result: Fail to reject H₀ → No significant difference in spending by gender.")
```

```python
print("\n[2] One-Way ANOVA: Sales Across Categories")
category_groups = [group['Total_Sales'].values for name, group in df.groupby('Category')]

f_stat, p_value = stats.f_oneway(*category_groups)

print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("Result: Reject H₀ → At least one category has significantly different sales.")
else:
    print("Result: Fail to reject H₀ → No significant difference across categories.")
```

```python
print("\n[3] Chi-Square Test: Gender vs Category Association")
contingency_table = pd.crosstab(df['Gender'], df['Category'])
chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2_stat:.4f}, P-Value: {p_val:.4f}, DoF: {dof}")
print("Significant association" if p_val < 0.05 else "No significant association")

print("\n" + "="*60)
print("SUMMARY FOR REPORT")
print("="*60)
print(f"• Gender vs Sales (t-test):        p = {stats.ttest_ind(male_sales, female_sales, equal_var=False)[1]:.4f}")
print(f"• Category vs Sales (ANOVA):       p = {f_stat:.4f} -> p-val = {stats.f_oneway(*category_groups)[1]:.4f}")
print(f"• Gender vs Category (Chi-Square): p = {chi2_stat:.4f} -> p-val = {stats.chi2_contingency(contingency_table)[1]:.4f}")
```
