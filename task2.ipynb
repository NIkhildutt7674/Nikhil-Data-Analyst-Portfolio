```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#creating the data frame
df=pd.read_csv("sales_data_cleaned.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Gender']=df['Gender'].astype("category")
df['City']=df['City'].astype("category")
df['Product']=df['Product'].astype("category")
df['Category']=df['Category'].astype("category")
df['Month']=df['Month'].astype("category")
df['Age_Group']=df['Age_Group'].astype("category")
df['Quarter']=df['Quarter'].astype("category")
df['Day_Name']=df['Day_Name'].astype("category")
```

```python
df.dtypes
```

```python
#univariate analysis
plt.hist(df['Total_Sales'],bins=10)
plt.title('Sales Distribution')
plt.show()
```

```python
#univariate Analysis
#numerical columns
num_col=df.select_dtypes(include = ['number'])
for num in num_col:
    print('column name:',num)
    print('description:',df[num].describe())
    sns.histplot(df[num], kde=True)
    plt.show()

    sns.boxplot(x=df[num])
    plt.show()
```

```python
#univariate Analysis
#categorical columns
cat_col=df.select_dtypes(include = ['category'])
for cat in cat_col:
    counts=df[cat].value_counts()
    print('value counts',counts)
    plt.bar(counts.index,counts)
    #sns.countplot(x=cat,data=df)
    plt.show()
```

Age_Group,Category is moderate balanced class

```python
#multivariate analysis
num_col=df.select_dtypes(include = ['number']).columns
print(df[num_col].corr())
corr=df[num_col].corr()
sns.heatmap(corr,annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.scatter(df['Quantity'],df['Total_Sales'])
plt.xlabel("Quantity")
plt.ylabel("Total Sales")
plt.show()
```

```python
sns.boxplot(x='Category',y='Total_Sales',data=df)
plt.show()
```
